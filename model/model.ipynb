```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2 = scaler.fit_transform(np.array(df1).reshape(-1,1))
```

```python
df2.shape
```

```python
df2
```

```python
training_set = int(len(df2)*0.65)
test_size = len(df2)-training_set
train_data,test_data = df2[0:training_set,:],df2[training_set:len(df2),:]
```

```python
training_set,test_size
```

```python
train_data
```

```python
def create_dataset(dataset,time_stamp=1):
    dataX,dataY = [],[]
    for i in range(len(dataset)-time_stamp-1):
        a = dataset[i:(i+time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 100
X_train , y_train = create_dataset(train_data,time_step)
X_test , y_test = create_dataset(test_data,time_step)
```

```python
print(X_train.shape),print(y_train.shape)
```

```python
print(X_test.shape),print(y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=64,verbose=1)
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
train_predict
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
math.sqrt(mean_squared_error(y_test,test_predict))
```

```python
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
```
